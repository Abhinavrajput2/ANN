```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Churn_Modelling (1).csv")
```

```python
df.head()
```

```python
X = df.iloc[:, 3:13]
X.head()
```

```python
y = df.iloc[:, 13]
y.head()
```

```python
geography=pd.get_dummies(X["Geography"],drop_first=True)
geography.head()
```

```python
gender=pd.get_dummies(X["Gender"],drop_first=True)
```

```python
X=pd.concat([X,geography,gender],axis=1)
```

```python
X.head()
```

```python
X=X.drop(['Geography','Gender'],axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
# how to see the no of column in x_train
len(x_train[0])
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
model=Sequential()
#Adding the input layer and the first hidden layer
model.add(Dense(units=6,kernel_initializer="he_uniform",activation="relu",input_dim=11))
```

```python
#Adding the second hidden layer
model.add(Dense(units=6,kernel_initializer="he_uniform",activation="relu"))
```

```python
#Adding the output layer
model.add(Dense(units=1,kernel_initializer="glorot_uniform",activation='sigmoid'))
```

```python
# compiling the ann model
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
fit_model=model.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=20)
```

```python
model.summary()
```

```python
y_pred = model.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix{}".format(cm))
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print("Accueacy score{}".format(score))
```
